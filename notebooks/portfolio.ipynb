```python
import pandas as pd
import yfinance as yf
# import matplotlib.pyplot as plt
# import seaborn as sns
import plotly.express as px
```

```python
df_portfolio = pd.read_csv(
    "data/20240803-portfolio.csv",
    index_col="Update Date",
    parse_dates=True,
    dayfirst=True,
)
```

```python
# Exchange Rate  # USD to THB
ticker = 'USDTHB=X'
data = yf.Ticker(ticker)

USDTHB = data.history(period='1d')['Close'].iloc[-1]
THBUSD = 1 / USDTHB

print("Exchange Rate\n-------------")
print(f"Date: {data.history(period='1d').index[0].date()}")
print(f"USDTHB: ฿{USDTHB :.2f}")
print(f"THBUSD: ${THBUSD :.4f}")
```

```python
df_portfolio["Total Cost (USD)"] = (
    df_portfolio["Total Cost"]
    .str.replace(",", "")
    .apply(lambda x: float(x[1:]) * THBUSD if x.startswith("฿") else float(x[1:]))
)

df_portfolio["Total Cost (THB)"] = (
    df_portfolio["Total Cost"]
    .str.replace(",", "")
    .apply(lambda x: float(x[1:]) * USDTHB if x.startswith("$") else float(x[1:]))
)

df_portfolio.sort_values(by="Total Cost (THB)", ascending=False, inplace=True)
```

```python
df_portfolio['Proportion (%)'] = df_portfolio['Total Cost (THB)'] / df_portfolio['Total Cost (THB)'].sum() * 100
```

```python
# bar chart
fig = px.bar(
    df_portfolio,
    x=df_portfolio["Ticker"],
    y="Total Cost (THB)",
    title="Total Cost (THB)",
    color="Total Cost (THB)",
)
fig.show()

# treemap
fig = px.treemap(
    df_portfolio,
    path=["Ticker"],
    values="Total Cost (THB)",
    title="Total Cost (THB)",
    color="Total Cost (THB)"
)
fig.show()
```
