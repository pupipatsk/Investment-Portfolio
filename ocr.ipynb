```python
import pandas as pd
# OCR
import pytesseract
pytesseract.pytesseract.tesseract_cmd = '/opt/homebrew/bin/tesseract'  # Path to tesseract executable  # For Apple Silicon Macs
from PIL import Image
import re # Regular expressions
```

```python
# Load an image
image = Image.open('data/transaction/20240331-B-NVDA.png')

# Perform OCR
text = pytesseract.image_to_string(image, lang='eng')

print(text)
```

```python
# # Given text
# text = '''Status (As of 31 Mar 2024 - 12:52)

# e Waiting

# —
# Your buy order is submitted and =e
# will be matched with sell orders e
# during regular trading hours after
# the currency exchange from THB
# to USD is completed.
# Buy NVDA NASDAQ
# 999.92 THB
# Stock Amount 999.92 THB
# Commission Fee 0.00 THB
# VAT 7% 0.00 THB
# Exchange Rate 1 USD = 36.48 THB
# USD Amount 27.41 USD
# Submission Date @ Dime! Fast

# 31 Mar 2024 - 12:52

# Order Type (s) Market Order
# Dime! Portfolio US Stock
# Offshore Account No. 80000223127
# Order ID STKBML20240331

# 0552153308625

# Payment Account

# Account Name uig AAWNS avus
# Account No. 202-583733-9
# Reference ID FX202403310552160519869

# Payment Ref ID OIHT5JTDV85MVV50YM4H57TOD4

# Receiving Account
# Account Name Dime! USD
# Account No. 800-002231-27

# Receiving Ref ID STKSWI20240331055218714736

# Purpose : Offshore Investment

# *The exchange rate is powered by Kiatnakin Phatra Bank. KKP
# Dime is only a channel for the currency exchange service by
# Kiatnakin Phatra Bank and not an authorized currency
# exchange operator.

# This order will be matched during regular trading
# hours at the best available price until you cancel it. It
# will be expired at the market close. ‘+
# '''

# Regular expressions to extract data
status_pattern = re.search(r'Status \(As of (.*?)\)', text)
status = status_pattern.group(1) if status_pattern else None

waiting_pattern = re.search(r'\ne (\w+)\n\n—', text)
waiting = waiting_pattern.group(1) if waiting_pattern else None

order_type_pattern = re.search(r'Order Type \(s\) (.*?)\n', text)
order_type = order_type_pattern.group(1) if order_type_pattern else None

portfolio_pattern = re.search(r'Dime! Portfolio (.*?)\n', text)
portfolio = portfolio_pattern.group(1) if portfolio_pattern else None

offshore_account_pattern = re.search(r'Offshore Account No. (.*?)\n', text)
offshore_account = offshore_account_pattern.group(1) if offshore_account_pattern else None

order_id_pattern = re.search(r'Order ID (.*?)\n\n', text)
order_id = order_id_pattern.group(1) if order_id_pattern else None

payment_account_name_pattern = re.search(r'Account Name (.*?)\nAccount No.', text)
payment_account_name = payment_account_name_pattern.group(1) if payment_account_name_pattern else None

payment_account_no_pattern = re.search(r'Account No. (.*?)\nReference ID', text)
payment_account_no = payment_account_no_pattern.group(1) if payment_account_no_pattern else None

reference_id_pattern = re.search(r'Reference ID (.*?)\n\n', text)
reference_id = reference_id_pattern.group(1) if reference_id_pattern else None

payment_ref_id_pattern = re.search(r'Payment Ref ID (.*?)\n\n', text)
payment_ref_id = payment_ref_id_pattern.group(1) if payment_ref_id_pattern else None

receiving_account_name_pattern = re.search(r'Receiving Account\nAccount Name (.*?)\nAccount No.', text)
receiving_account_name = receiving_account_name_pattern.group(1) if receiving_account_name_pattern else None

receiving_account_no_pattern = re.search(r'Receiving Account\n.*\nAccount No. (.*?)\n\n', text)
receiving_account_no = receiving_account_no_pattern.group(1) if receiving_account_no_pattern else None

receiving_ref_id_pattern = re.search(r'Receiving Ref ID (.*?)\n\n', text)
receiving_ref_id = receiving_ref_id_pattern.group(1) if receiving_ref_id_pattern else None

purpose_pattern = re.search(r'Purpose : (.*?)\n\n', text)
purpose = purpose_pattern.group(1) if purpose_pattern else None

# Organize extracted data into a dictionary
data = {
    "Status": [status],
    "Waiting": [waiting],
    "Order Type": [order_type],
    "Portfolio": [portfolio],
    "Offshore Account No.": [offshore_account],
    "Order ID": [order_id],
    "Payment Account Name": [payment_account_name],
    "Payment Account No.": [payment_account_no],
    "Reference ID": [reference_id],
    "Payment Ref ID": [payment_ref_id],
    "Receiving Account Name": [receiving_account_name],
    "Receiving Account No.": [receiving_account_no],
    "Receiving Ref ID": [receiving_ref_id],
    "Purpose": [purpose]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display DataFrame
df
```
